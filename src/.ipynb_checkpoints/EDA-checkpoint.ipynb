```python
train_city_df = train_df[['City','lat','lon']].drop_duplicates().reset_index(drop=True)
train_city_df
```

```python
test_city_df = test_df[['City','lat','lon']].drop_duplicates().reset_index(drop=True)
test_city_df
```

```python
def city_distance(query, key, train_mode=False):
    np_query = np.array(query[["lat", "lon"]])
    np_key = np.array(key[["lat", "lon"]])
    nearest_indexs = []
    nearest_cities = []
    nearest_distances = []
    nearest_lats = []
    nearest_lons = []
    for q in np_query:
        distance = np.sum(np.square(np_key-q), axis=1)
        if train_mode:
            nearest_index = np.argsort(distance)[1]
        else:
            nearest_index = np.argmin(distance)
        nearest_indexs.append(nearest_index)
        nearest_cities.append(key["City"][nearest_index])
        nearest_distances.append(distance[nearest_index])
        nearest_lat, nearest_lon = np_key[nearest_index]
        nearest_lats.append(nearest_lat)
        nearest_lons.append(nearest_lon)
    # print(nearest_indexs)
    # print(nearest_cities)
    query["City_nearest"] = nearest_cities
    query["City_distance_nearest"] = nearest_distances
    query["City_lat_nearest"] = nearest_lats
    query["City_lon_nearest"] = nearest_lons
    
    return query.drop(["lat", "lon"], axis=1)
```

```python
train_nearest_city_df = city_distance(train_city_df, train_city_df, train_mode=True)
```

```python
test_nearest_city_df = city_distance(test_city_df, train_city_df)
```

```python
train_nearest_city_df
```

```python
new_train_df = pd.merge(train_df, train_nearest_city_df, on="City")
new_train_df
new_train_df.info()
```

```python
new_test_df = pd.merge(test_df, test_nearest_city_df, on="City")
new_test_df
new_test_df.info()
```

```python
new_new_train_df = pd.merge(new_train_df, new_train_df[["year", "month", "day", "City", "pm25_mid"]], left_on=["year", "month", "day", "City_nearest"], right_on=["year", "month", "day", "City"], how="left", suffixes=['', '_nearest'])
new_new_train_df = new_new_train_df.loc[:,~new_new_train_df.columns.duplicated()].fillna(method='ffill').sort_values('id')
new_new_train_df.info()
```

```python
new_new_test_df = pd.merge(new_test_df, new_train_df[["year", "month", "day", "City", "pm25_mid"]], left_on=["year", "month", "day", "City_nearest"], right_on=["year", "month", "day", "City"], how="left", suffixes=['', '_nearest'])
new_new_test_df = new_new_test_df.loc[:,~new_new_test_df.columns.duplicated()].fillna(method='ffill').sort_values('id')
new_new_test_df.info()
```
